```python
# 1

def DicNumAlpha(*argv):
  numeros = []
  textos = []
  for arg in argv:
    if type(arg)==int: #Para saber si es un dato de tipo entero
      numeros.append(arg) #Agrego a la lista de numeros
    elif type(arg) == str: #Para saber si arg es tipo string
      if arg.isnumeric(): #Si el string es numerico
        numeros.append(arg)
      else:
        textos.append(arg)
  return {'text':textos, 'num':numeros}
```

```python
# 2
def factorial(n: int) -> int:
  fac = 1
  for k in range(2,n+1):
    fac = k*fac
  return fac
```

```python
def factorialr(n: int) -> int:
  if n == 0:
    return 1
  else:
    return n*factorialr(n-1)
```

```python
def tail_factorial(n, accumulator=1):
  if n == 0: 
    return accumulator
  else: 
    return tail_factorial(n-1, accumulator * n)
```

```python
# 4  Del diccionario anterior escribir una funcion donde se filtren las emisoras 
#desde un limite inferior y un limite superior, retorna un diccionario solo 
#con los datos de las emisoras que cumplan con las condiciones.


def filtraEmi(FMdic, limInf, limSup): #104.0-107.0 -> Estrella y Olimpica
  dicOut = {}
  flag = True
  for indiceKey,freqVal in FMdic['Frecuencia (MHz)'].items():
    #indiceKey: sub llaves dentro de Frecuencia, del 0 al 33 (enteros)
    if (freqVal <= limSup) and (freqVal >= limInf):
      if flag: #para inicializar las llaves de dicOut
        for keyOr in FMdic.keys():
          dicOut[keyOr] = {indiceKey: FMdic[keyOr][indiceKey]} #se crean las llaves por primera vez
        flag = False #para la siguiente frecuencia ya se crearon las llaves
      else:
        for keyOr in FMdic.keys():
          dicOut[keyOr][indiceKey] = FMdic[keyOr][indiceKey] #asignamos los valores

  return dicOut


```

```python
#5  Escribir una funcion que reciba una cantidad finita de diccionarios 
#   y retorne la concatenacion de los diccionarios. Asumimos que tienen
# diferentes llaves

def concatDic(*argv):
  dicOut = argv[0].copy()
  for dic in argv[1:]:
    for key,val in dic.items():
      if key in dicOut.keys():
        dicOut[key] = [dicOut[key],dic[key]]
      else:
        dicOut[key] = val
  return dicOut

```

```python
#6
def sumaLlaves(fun,dic1, dic2):
  dicOut = dic1.copy()
  for key,val in dic2.items():
    if key in dicOut.keys():
      dicOut[key] = fun(dicOut[key],dic2[key])
    else:
      dicOut[key] = val
  return dicOut
```

```python
d1 = {'a': 50, 'b': 100, 'c':500}
d2 = {'a': 70, 'b': 400, 'd':400}
def suma(x,y): return x+y

salida = sumaLlaves(suma, d1, d2)
salida
```

```python
#7
#  Escribir una funcion que reciba una lista con caracteres, enteros, reales, 
# complejos, listas y tuplas. Retornar un diccionario que agrupe cada tipo de 
# dato en una lista, por ejemplo, todas los enteros agruparlos en un lista con 
# llave ”Enteros”.

def agrupaEnDic(lista):
  dic = {}
  nombresLlaves = {int:'Enteros',str:'Textos', complex:'Complejos',
                   tuple:'Tuplas', list:'Listas',float:'Flotantes'}
  for elem in lista:
    nkey = nombresLlaves[type(elem)]
    if dic.get(nkey,0):
      dic[nkey].append(elem)
    else:
      dic[nkey] = [elem]
  return dic
```

```python
dic = agrupaEnDic([1,(2,3),[4,5,6],7.0,8.0+9j,'10',[3,4],9.0,1,2-1j,'Hola'])
dic
```
