## Analicemos

La definición formal de un número aleatorio en un rango es: $(b-a)*rand + a$. Utilicemos esta definición para crear una matriz aleatoria cuyos elementos tengan valores entre 5 y 15.

```python
import random
#seed o semilla
M = np.zeros([5,5])

a = 5
b = 15

for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        M[i,j] = (b-a)*random.random()+a
print(M)
```


# Funciones vectorizadas

Numpy contiene las versiones vectorizadas de las funciones de la librería math, por ejemplo: 
- math.sin --> numpy.sin
- math.exp --> numpy.exp

A diferencia de las funciones de la librería math, las versiones vectorizadas pueden recibir arreglos n-dimensionales como argumentos.

```python
import math

def sigmoid(z):
    return 1. / (1. + math.exp(-z))
```

```python
z = np.linspace(-10,10,21)
print(z)
print(z.shape)
```

```python
sigmoid(z)
```

```python
R = []
for i in z:
    R.append(sigmoid(i))

plt.plot(z,R)
plt.grid()
```

```python
help(np.vectorize)
```

```python
sigmoid_vec = np.vectorize(sigmoid)
```

```python
type(sigmoid_vec)
```

```python
z = np.linspace(-10,10,21)
print(z)
y = sigmoid_vec(z)

print(y)
plt.plot(z,y,'k-',z,R,'r--')
plt.grid()
plt.xlim([-5,5])
```

```python
def sigmoid(z):
    return 1. / (1. + math.exp(-z))
    
def sigmoide_np(z):
    return 1. / (1. + np.exp(-z))
```

```python
Snp = sigmoide_np(z)
print(Snp.shape)
plt.plot(z,Snp)
plt.grid()
```

```python
SigmoideLambda = lambda z: 1. / (1. + math.exp(-z))
#funcional: map, filter, reduce
```

```python
%%timeit 
list(map(sigmoid,z))
```

```python
%%timeit
y2 = list(map(sigmoid, z))
```

```python
y2
```

```python
%%timeit
sigmoide_np(z)
```

```python
type(SigmoideLambda)
```

```python
SL = SigmoideLambda(z)

plt.plot(z,SL)
plt.grid()
```

Supongamos dos arreglos A y B de tamaño $m \times 2$, donde $m$ es un número de puntos y $2$ es la dimensión de los puntos, es decir, 2D para este ejemplo.

Vectorizar la distancia Euclídea entre cada punto 2D en A y su correspondiente punto 2D en B. El resultado debe ser un arreglo de dimensión $m$.

```python
np.random.seed(2)

m = 4

A = np.random.rand(m, 2)
B = np.random.rand(m, 2)
```

```python
A
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(3,3))
plt.plot(A[:,0], A[:,1], 'r*')
plt.plot(B[:,0], B[:,1], 'b*')
plt.show()
```

```python
def distAB(ax,ay,bx,by):
  return math.sqrt((ax-bx)**2+(ay-by)**2)
```

```python
R = []
for k in range(A.shape[0]):
  R.append(distAB(A[k,0],A[k,1],B[k,0],B[k,1]))
```

```python
R
```

```python
distAB_vec = np.vectorize(distAB)
```

```python
distAB_vec(A[:,0],A[:,1],B[:,0],B[:,1])
```
